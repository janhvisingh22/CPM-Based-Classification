```python
import json
```

```python
def read_large_json(jsonl_path, num_samples=3):
    with open(jsonl_path, "r", encoding="utf-8") as file:
        for i, line in enumerate(file):
            if i >= num_samples:
                break
            print(json.loads(line))  # Read each line as a JSON object

read_large_json("data/processed/NYT_preprocessed.jsonl")
```
